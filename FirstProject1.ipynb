```python
def moving(y):
    for i in range(y):
        print("moving forward..")
```

```python
import random
```

```python
elements = [["Tringle","Red"],["Circle","Blue"],["Tringle","Green"],["Square","Red"],["Circle","Red"],["Square","Green"],["Tringle","Blue"],["Circle","Green"],["Square","Blue"]]
```

```python
 Tringle=["r","g","b"]
Circle=["r","g","b"]
Square=["r","g","b"]
```

```python
s=input("enter start")
if s=="start":
    i=0
    while (elements):
    
        x=random.choice(elements)
        i=i+1
        print(i)
        elements.remove(x)
        if x[0]=="Tringle":
            moving(12)
            if x[1]=="Red":
                Tringle.insert(0,x)
                Tringle.remove("r")
            elif x[1]=="Green":
                Tringle.insert(1,x)
                Tringle.remove("g")
            else:
                Tringle.insert(2,x)
                Tringle.remove("b")
            print("it was a Tringle")
            moving(18)
            start()
        elif x[0]=="Circle":
            moving(15)
            if x[1]=="Red":
                Circle.insert(0,x)
                Circle.remove("r")
            elif x[1]=="Green":
                Circle.insert(1,x)
                Circle.remove("g")
            else:
                Circle.insert(2,x)
                Circle.remove("b")
            print("it was a Circle")
            moving(15)
            start()
        elif x[0]=="Square":
            moving(19)
            if x[1]=="Red":
                Square.insert(0,x)
                Square.remove("r")
            elif x[1]=="Green":
                Square.insert(1,x)
                Square.remove("g")
            else:
                Square.insert(2,x)
                Square.remove("b")
            print("it was a Square")
            moving(11)
            start()
        else:
            print("something went wrong...")
```
